## 1 - Cleaning and Preparing Data in Python

```python
from csv import reader
opened_file = open('artworks.csv')
read_file = reader(opened_file)
moma = list(read_file)
opened_file.close()
```

```python
# For explain purposes
# You can replace parts of strings using str.replace
age1 = "I am thirty-one years old"
age2 = age1.replace('one','two')
print(age2)
```

```python
# Column 3 (Nationality) and column 6 (Gender) have a string beetwen parentheses
# So let's remove it
for row in moma:
    row[2] = row[2].replace("(","")
    row[2] = row[2].replace(")","")
    
    row[5] = row[5].replace("(","")
    row[5] = row[5].replace(")","")
```

```python
# Handling missing values on columns 2 and 5
for row in moma:
    gender = row[5]
    gender = gender.title()
    if not gender:
        gender = "Gender Unknown/Other"
    row[5] = gender
    
    nationality = row[2]
    nationality = nationality.title()
    if not nationality:
        nationality = "Nationality Unknown"
    row[2] = nationality
```

```python
# Handling dates on columns 4 and 5
def clean_and_convert(date):
    # check that we don't have an empty string
    if date != "":
        # move the rest of the function inside
        # the if statement
        date = date.replace("(", "")
        date = date.replace(")", "")
        date = int(date)
    return date

for row in moma[1:]:
    begin = row[3]
    end = row[4]
    
    begin = clean_and_convert(begin)
    end = clean_and_convert(end)
    
    row[3] = begin
    row[4] = end
```

```python
# For explain purposes
# You can create a function to strip characters

bad_chars = ["(",")","c","C",".","s","'", " "]
test_data = ["1912", "1929", "1913-1923",
             "(1951)", "1994", "1934",
             "c. 1915", "1995", "c. 1912",
             "(1988)", "2002", "1957-1959",
             "c. 1955.", "c. 1970's", 
             "C. 1990-1999"]

def strip_characters(string):
    for bc in bad_chars:
        string = string.replace(bc, "")
    return string

stripped_test_data = []
for td in test_data:
    td = strip_characters(td)
    stripped_test_data.append(td)
```

```python
# You can create a function to handle range data
def process_date(string):
    if "-" in string:
        first_year = int(string[:4])
        end_year = int(string[5:])
        avg = round((first_year  + end_year)/2)
        return avg
    else:
        return int(string)
    
processed_test_data = []
for std in stripped_test_data:
    std = process_date(std)
    processed_test_data.append(std)
    
for row in moma:
    date = row[6]
    date = strip_characters(date)
    date = process_date(date)
    row[6] = date
```

```python
# You can create customized classes and create methods

class NewList():
    """
    A Python list with some extras!
    """
    def __init__(self, initial_state):
        self.data = initial_state
        
    def append(self, new_item):
        """
        Append `new_item` to the NewList
        """
        self.data = self.data + [new_item]
        self.calc_length()
    
    def calc_length(self):
        self.length = len(self.data)
        
fibonacci = NewList([1,1,2,3,5])
fibonacci.append(8)
print(fibonacci.length)
```

## 3 - Working with Dates and Times in Python

```python
from csv import reader
opened_file = open('potus_visitors_2015.csv')
read_file = reader(opened_file)
potus = list(read_file)
potus = potus[1:]
opened_file.close()
```

```python
# For explain purposes
# You can formate a date using datetime

import datetime as dt
ibm_founded = dt.datetime(1911, 6, 16)
man_on_moon = dt.datetime(1969, 7, 20, 20, 17)
```

```python
# Let's formate the dates on column 3 as datetime

date_format = "%m/%d/%y %H:%M"

for row in potus:
    start_date = row[2]
    start_date = dt.datetime.strptime(start_date, date_format)
    row[2] = start_date
```

```python
# With this you can calculate the number of visitors per month

visitors_per_month ={}

import datetime as dt

for row in potus:
    value = row[2]
    string = value.strftime("%B, %Y")
    if string in visitors_per_month:
        visitors_per_month[string] += 1
    else:
        visitors_per_month[string] = 1
```

```python
# For explain purposes
# You can get the first time with 'min()' and last time with 'max()'

appt_times = []

for row in potus:
    datetime_object = row[2]
    new_time = datetime_object.time()
    appt_times.append(new_time)

min_time = min(appt_times)
max_time = max(appt_times)

print(min_time)
print(max_time)
```

```python
# For explain purposes
# you can do operations using timedelta

dt_1 = dt.datetime(1981, 1, 31)
dt_2 = dt.datetime(1984, 6, 28)
dt_3 = dt.datetime(2016, 5, 24)
dt_4 = dt.datetime(2001, 1, 1, 8, 24, 13)

answer_1 = dt_2 - dt_1
answer_2 = dt_3 + dt.timedelta(days=56)
answer_3 = dt_4 - dt.timedelta(seconds=3600)

print(answer_1)
print(answer_2)
print(answer_3)
```

```python
# Let's work on dataset

for row in potus:
    end_date = row[3]
    end_date = dt.datetime.strptime(end_date, "%m/%d/%y %H:%M")
    row[3] = end_date
    
appt_lengths = {}

for row in potus:
    start_date = row[2]
    end_date = row[3]
    
    length = end_date - start_date
    if length in appt_lengths:
        appt_lengths[length] += 1
    else:
        appt_lengths[length] = 1
        
min_length = min(appt_lengths)
max_length = max(appt_lengths)
print(min_length)
print(max_length)
```
